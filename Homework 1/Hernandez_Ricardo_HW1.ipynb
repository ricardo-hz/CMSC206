```python
#Author: Ricardo Hernandez
#Program: Celsius to Fahrenheit & Vice Versa, LB to KG & Vice Versa

POUNDS_PER_KILOGRAM = 2.20462262185
#Used for the continuation of the program
convertAnother = 'Y'

def celsiusToFahrenheit(number):
  print("{} C = {} F".format(number, ((number * (9/5) + 32))))

def fahrenheitToCelsius(number):
  print("{} F = {} C".format(number, ((number - 32) / (9/5))))

def poundsToKilograms(number):
  print ("{} lbs = {} kg".format(number, (number/POUNDS_PER_KILOGRAM)))

def kilogramsToPounds(number):
  print("{} kg = {} lbs".format(number, (number * POUNDS_PER_KILOGRAM)))

#ENHANCEMENT: Allow for multiple conversions in one run
while (convertAnother == 'Y' or convertAnother == 'y'):
  #Prompt for conversion
  print("Choose a conversion by entering a number 1-4")
  print(" 1: Celsius to Fahrenheit")
  print(" 2: Fahrenheit to Celsius")
  #ENHANCEMENT: Allow for conversion from lbs to kg and vice versa
  print(" 3: Pounds to Kilograms")
  print(" 4: Kilograms to Pounds")
  #Validate entry
  conversionChoice = int(input())
  while(conversionChoice < 1 or conversionChoice > 4):
    print("Please choose a valid conversion by entering a number 1-4")
    conversionChoice = int(input())

  #Prompt for number to convert
  print("\nEnter a number to convert")
  numToConvert = float(input())

  #Begin conversions
  #Celsius to Fahrenheit
  if (conversionChoice == 1):
    celsiusToFahrenheit(numToConvert)
  #Fahrenheit to Celsius
  elif (conversionChoice == 2):
    fahrenheitToCelsius(numToConvert)
  #Pounds to Kilograms
  elif(conversionChoice == 3):
    poundsToKilograms(numToConvert)
  #Kilograms to Pounds. Catch all case if user bypasses input verification
  else:
    kilogramsToPounds(numToConvert)
  print("\nWould you like to convert another number? Y/N")
  convertAnother = input()
```
